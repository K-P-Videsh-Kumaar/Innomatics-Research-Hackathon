```python
#importing pandas
import pandas as pd
import numpy as np
import matplotlib as plt
import sqlite3
```

```python
#orders data
df=pd.read_csv("orders.csv")
df1=pd.read_json('users.json')
df2=pd.read_excel('Res_data.xlsx')
print()
```

```python
#Merging the data 
df_mer1=pd.merge(df,df1,on='user_id',how='left')
df_mer2=pd.merge(df_mer1,df2,on='restaurant_id',how='left')
df_mer2
```

## Which city has the highest total revenue (total_amount) from Gold members?

```python
df_mer2[df_mer2['membership'] == 'Gold'].groupby('city')['total_amount'].sum()
```

## Which cuisine has the highest average order value across all orders?

```python
df_mer2.groupby('cuisine')['total_amount'].mean()
```

## How many distinct users placed orders worth more than ₹1000 in total (sum of all their orders)?

```python
(df_mer2.groupby('user_id')['total_amount'].sum()>1000).sum()
```

## Which restaurant rating range generated the highest total revenue?

```python
df_pr=df_mer2.groupby('rating')['total_amount'].sum().sort_values()
```

```python
count1=count2=count3=count4=0
for i,j in df_pr.items():
    if i>=3.0 and i<=3.5:
        count1+=j
    elif i>=3.6 and i<=4.0:
        count2+=j
    elif i>=4.1 and i<=4.5:
        count3+=j
    elif i>=4.6 and i<=5.0:
        count4+=j
print(count1,count2,count3,count4)
```
